## Part 1 of Simplified Vending Machine

```python
menu = [
       'Soft Drink', 's', '$3.00',
       'Iced Coffee', 'c', '$4.00',
       'Enery Drink', 'e', '$3.50',
       'Water', 'w', '$2.00',
       'Green Tea', 't', '$2.50']

import numpy as np
np.array(menu)

np.reshape(menu,(5,3)) 

from tabulate import tabulate
col_names = ['Product', 'Symbol', 'Price']
print(tabulate(np.reshape(menu,(5,3)) , headers=col_names))

symbol = (input('Please choose the symbol: '))

dict = {0.5 : '50 cents', 1.00 : '1 dollar', 1.50 : '1 dollar and 50 cents', 
        2.00 : '2 dollars', 2.50 : '2 dollars and 50 cents', 3.00 : '3 dollars',
        3.50 : '3 dollars and 50 cents'}

change = {0.5: '1 fifty-cents', 1 : '1 one dollar', 1.5 : '1 one dollar & 1 fifty-cents', 2 : '1 two dollar',
          2.5: '1 two dollar & 1 fifty-cents'}


if symbol == 's':
    print('The price of Soft Drink is $3.00')
    coin = float(input('Please insert a coin of 50c, $1, or $2: '))
    soft_drink = 3.00
    k = dict.get(soft_drink - coin)
    j = change.get(coin - soft_drink)
    
    while coin < 3:
         print('The outstanding amount is $',k  )
         
         balance = float(input('Please insert a coin of 50c, $1, or $2: '))
         coin = coin + balance
         k = dict.get(soft_drink - coin)
         j = change.get(coin - soft_drink)
        
    if coin > 3:
        print('Change: $', coin - soft_drink)
        
        print('Please take the following change: ')
        
        print(j)
        
    else:
            print('You paid the exact amount, Thanks')
            
    
        
elif symbol == 'c':
    print('The price of Iced Coffee is $4.00')
    coin = float(input('Please insert a coin of 50c, $1, or $2: '))
    iced_coffee = 4.00
    a = dict.get(iced_coffee - coin)
    b = change.get(coin - iced_coffee)
    
    while coin < 4:
         print('The outstanding amount is $',a )
         
         balance = float(input('Please insert a coin of 50c, $1, or $2: '))
         coin = coin + balance
         a = dict.get(iced_coffee - coin)
         b = change.get(coin - iced_coffee)
    
    
    if coin > 4:
        print('Change: $', coin - iced_coffee)
        
        print('Please take the following change: ')
        
        print(b)
        
    else:
            print('You paid the exact amount, Thanks')



elif symbol == 'e':
    print('The price of Energy Drink is $3.50')
    coin = float(input('Please insert a coin of 50c, $1, or $2: '))
    energy_drink = 3.50
    g = dict.get(energy_drink - coin)
    h = change.get(coin - energy_drink)
    
    while coin < 3.50:
         print('The outstanding amount is $',g  )
         
         balance = float(input('Please insert a coin of 50c, $1, or $2: '))
         coin = coin + balance
         g = dict.get(energy_drink - coin)
         h = change.get(coin - energy_drink)
    
    if coin > 3.50:
        print('Change: $', coin - energy_drink)
        
        print('Please take the following change: ')
        
        print(h)
    else:
            print('You paid the exact amount, Thanks')
    
  
    
    
elif symbol == 'w':
    print('The price of Water is $2.00')
    coin = float(input('Please insert a coin of 50c, $1, or $2: '))
    water = 2.00
    l = dict.get(water - coin)
    m = change.get(coin - water)
    while coin < 2:
         print('The outstanding amount is $',l  )
         
         balance = float(input('Please insert a coin of 50c, $1, or $2: '))
         coin = coin + balance
         l = dict.get(water - coin)
         m = change.get(coin - water)
    
    if coin > 2:
        print('Change: $', coin - water)
        
        print('Please take the following change: ')
        
        print(m)
        
    else:
            print('You paid the exact amount, Thanks')
    
    
    
    
elif symbol == 't':
    print('The price of Green Tea is $2.50')
    coin = float(input('Please insert a coin of 50c, $1, or $2: '))
    green_tea = 2.50
    p = dict.get(green_tea - coin)
    q = change.get(coin - green_tea)
    while coin < 2.50:
         print('The outstanding amount is $', p )
         
         balance = float(input('Please insert a coin of 50c, $1, or $2: '))
         coin = coin + balance
         p = dict.get(green_tea - coin)
         q = change.get(coin - green_tea)
    if coin > 2.50:
       print('Change: $', coin - green_tea)

       print('Please take the following change: ')
        
       print(q)
        
    else:
            print('You paid the exact amount, Thanks')
            
            
    
    



```
